```python
food_data['dateAdded'] = pd.to_datetime(food_data['dateAdded'])
food_data['dateUpdated'] = pd.to_datetime(food_data['dateUpdated'])

food_data['dateAdded'] = food_data['dateAdded'].dt.strftime('%m-%d-%Y')
food_data['dateUpdated'] = food_data['dateUpdated'].dt.strftime('%m-%d-%Y')

food_data.head()
```

### Task 1E: 5 points
For the highest fast food crowded city, display a pie chart of top 10 distinct restaurants with percentages.

### Task 1F: 5 points

1. Print Top 10 highest count of Ice cream shops and their shop name for each province.
2. Print Top 10 highest count of Sandwich shops and their shop name for each province.
3. Display two horizontal bar graph side to each other for both the results with appropriate titles and labels.

```python
ice_cream_shops = food_data[food_data['categories'].str.contains('Ice Cream', case=False, na=False)]
top_ice_cream_shops_by_province = ice_cream_shops.groupby('province')['name'].value_counts().groupby(level=0).head(10)
print(top_ice_cream_shops_by_province)
```

```python
sandwich_shops = food_data[food_data['categories'].str.contains('Sandwich', case=False, na=False)]
top_sandwich_shops_by_province = sandwich_shops.groupby('province')['name'].value_counts().groupby(level=0).head(10)
print(top_sandwich_shops_by_province)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 8))

# Top ice cream shops plot
top_ice_cream_shops_by_province.unstack().fillna(0).sum(axis=1).sort_values().tail(10).plot(kind='barh', ax=axes[0], color='skyblue')
axes[0].set_title('Top 10 Provinces by Ice Cream Shop Count')
axes[0].set_xlabel('Number of Ice Cream Shops')
axes[0].set_ylabel('Province')

# Top sandwich shops plot
top_sandwich_shops_by_province.unstack().fillna(0).sum(axis=1).sort_values().tail(10).plot(kind='barh', ax=axes[1], color='lightgreen')
axes[1].set_title('Top 10 Provinces by Sandwich Shop Count')
axes[1].set_xlabel('Number of Sandwich Shops')
axes[1].set_ylabel('Province')

plt.tight_layout()
plt.show()
```

### Task 1I: 5 points
1. Print no of added restaurants for years 2015, 2016 and 2017 for each distinct city.(including 0 counts)
2. Display a line chart showing each cities growth of 15 or more from years 2015 to 2017.

```python
#code referenced from chatGpt prompt:Print no of added restaurants for years 2015, 2016 and 2017 for each distinct city.(including 0 counts)
food_data['yearAdded'] = food_data['dateAdded'].apply(lambda x: x[-4:])

filtered_data = food_data[food_data['yearAdded'].isin(['2015', '2016', '2017'])]

restaurant_counts = filtered_data.groupby(['city', 'yearAdded']).size().unstack(fill_value=0)

all_cities = food_data['city'].unique()
missing_cities = set(all_cities) - set(restaurant_counts.index)
missing_data = pd.DataFrame(0, index=list(missing_cities), columns=['2015', '2016', '2017'])

complete_data = pd.concat([restaurant_counts, missing_data]).fillna(0)

print(complete_data)
```

```python
#code referenced from chatGpt, prompt:Display a line chart showing each cities growth of 15 or more from years 2015 to 2017.
import matplotlib.pyplot as plt

growth = complete_data['2017'] - complete_data['2015']

cities_with_growth = growth[growth >= 15].index

plt.figure(figsize=(10, 6))

for city in cities_with_growth:
    plt.plot(complete_data.columns, complete_data.loc[city], marker='o', label=city)

plt.title('Growth of Restaurants from 2015 to 2017 by City')
plt.xlabel('Year')
plt.ylabel('Number of Added Restaurants')
plt.xticks(['2015', '2016', '2017'])
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Task 3B: 3 points
Display concavity worst vs smoothness worst with reaction to target variable.
Interpret the graph.

```python
# Plotting concavity_worst vs smoothness_worst with hue for diagnosis
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='concavity_worst', y='smoothness_worst', hue='diagnosis', style='diagnosis', palette='bright')
plt.title('Concavity Worst vs Smoothness Worst by Diagnosis')
plt.xlabel('Concavity Worst')
plt.ylabel('Smoothness Worst')
plt.legend(title='Diagnosis')
plt.show()
```
